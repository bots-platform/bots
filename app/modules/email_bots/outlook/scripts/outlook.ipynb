```python
df = df[df['Ticket'] == 'ok']
df.head()
```

```python

def extract_all_6_digits(text):
    return re.findall(r'\b[1-3]\d{7}\b', text)
```

```python
df['Numeros_6_digitos'] = df['Cuerpo'].apply(extract_all_6_digits)
df.head(50)
```

```python


number_pattern = re.compile(r'\b[1-3]\d{7}\b')
                            
sent_date_pattern = re.compile(
    r'(?:Enviado el:|Sent|Enviado)\s*(.+?)\s(\d{1,2}:\d{2}\s*(?:AM|PM|am|pm)?)',
    re.IGNORECASE
)

def extract_numbers_before_date(cuerpo):
    sent_date_match = sent_date_pattern.search(cuerpo)
    if sent_date_match:
        text_before_date = cuerpo[:sent_date_match.start()]
        numbers = number_pattern.findall(text_before_date)
    else:
        numbers = []
    return numbers
```

```python
df['tickets_encontrados'] = df['Cuerpo'].apply(extract_numbers_before_date)
fecha_inicio = "2025-01-25"
fecha_fin = "2025-01-25"

df['Fecha'] = pd.to_datetime(df['Fecha'])
df_filtrado = df[(df['Fecha'] >= fecha_inicio) & (df['Fecha'] <= fecha_fin) ]
df = df.sort_values(by="Fecha", ascending=False)
df_filtrado.head(100)
```

```python
df_filtrado = df[df['tickets_encontrados'].apply(lambda x : len(x) > 0 )]
df_filtrado.head()
```

```python
df_sorted = df_filtrado.sort_values(by="Fecha", ascending=True)
df_sorted.head()
```

```python
df_duplicados = df_sorted[df_sorted.duplicated(subset='tickets_encontrados', keep=False)]
print("Duplicados")
df_duplicados
```
